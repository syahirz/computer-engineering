```python
from IPython.display import Image, display  

#Setup
decimal_point = 3 # Decimal point for rounding
carry_over_tolerance = 0.001 # Carry over tolerance stopper
max_iterations = 100 # Max_Imum iteration

# Simple Beam Calculator
print('3-Span Moment Distribution Calculator')

# Diagram for referrence
display(Image(url='https://github.com/syahirz/computer-engineering/blob/main/1.jpg?raw=true', width=400))  

# Load Values Input  
print('Load Values (in kN/m for distributed loads, kN for point loads):')  

# Input for a_Load  
a_Load = float(input('Enter a value (a_Load) in kN/m: '))  
while a_Load <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    a_Load = float(input('Enter a value (a_Load) in kN/m: '))  

# Input for b_Load  
b_Load = float(input('Enter b value (b_Load) in kN/m: '))  
while b_Load <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    b_Load = float(input('Enter b value (b_Load) in kN/m: '))  

# Input for c_Load  
c_Load = float(input('Enter c value (c_Load) in kN/m: '))  
while c_Load <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    c_Load = float(input('Enter c value (c_Load) in kN/m: '))  

# Input for d_Load  
d_Load = float(input('Enter d value (d_Load) in kN: '))  
while d_Load <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    d_Load = float(input('Enter d value (d_Load) in kN: '))  

# Distance Values Input  
print('')
print('Distance Values (in m):')  

# Input for x_Distance  
x_Distance = float(input('Enter x value (x_Distance) in m: '))  
while x_Distance <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    x_Distance = float(input('Enter x value (x_Distance) in m: '))  

# Input for y_Distance  
y_Distance = float(input('Enter y value (y_Distance) in m: '))  
while y_Distance <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    y_Distance = float(input('Enter y value (y_Distance) in m: '))  

# Input for a_Distance  
a_Distance = float(input('Enter a value (a_Distance) in m: '))  
while a_Distance <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    a_Distance = float(input('Enter a value (a_Distance) in m: '))  

# Input for b_Distance  
b_Distance = float(input('Enter b value (b_Distance) in m: '))  
while b_Distance <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    b_Distance = float(input('Enter b value (b_Distance) in m: '))  

# Span Cross Section Values Input  
print('')
print('Span Cross Section Values (in mm^4):')  

# Input for x_I  
x_I = float(input('Enter 1st span I value (x_I) in mm^4: '))  
while x_I <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    x_I = float(input('Enter 1st span I value (x_I) in mm^4: '))  

# Input for y_I  
y_I = float(input('Enter 2nd span I value (y_I) in mm^4: '))  
while y_I <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    y_I = float(input('Enter 2nd span I value (y_I) in mm^4: '))  

# Input for z_I  
z_I = float(input('Enter 3rd span I value (z_I) in mm^4: '))  
while z_I <= 0:  
    print('Invalid input. Please enter a positive numeric value.')  
    z_I = float(input('Enter 3rd span I value (z_I) in mm^4: '))  


# Calculate Fixed End Moments (FEMs)  
FEM_AB = round(-((a_Load * x_Distance ** 2) / 12), decimal_point)  
FEM_BA = round((a_Load * x_Distance ** 2) / 12, decimal_point)  

if b_Load < c_Load:
    FEM_BC = round(-((b_Load * y_Distance ** 2 / 12) + ((c_Load - b_Load) * y_Distance ** 2 / 30)), decimal_point)  
    FEM_CB = round((b_Load * y_Distance ** 2 / 12) + ((c_Load - b_Load) * y_Distance ** 2 / 20), decimal_point)  
elif b_Load > c_Load:
    FEM_BC = round(-((c_Load * y_Distance ** 2 / 12) + ((b_Load - c_Load) * y_Distance ** 2 / 20)), decimal_point)  
    FEM_CB = round((c_Load * y_Distance ** 2 / 12) + ((b_Load - c_Load) * y_Distance ** 2 / 30), decimal_point) 

FEM_CD = round(-((d_Load * b_Distance ** 2 * a_Distance) / (a_Distance + b_Distance) ** 2), decimal_point)  
FEM_DC = round(((d_Load * a_Distance ** 2 * b_Distance) / (a_Distance + b_Distance) ** 2), decimal_point)  

# Calculate stiffness factors  
k_AB = 4 * x_I / x_Distance  
k_BA = 4 * x_I / x_Distance  
k_BC = 4 * y_I / y_Distance  
k_CB = 4 * y_I / y_Distance  
k_CD = 4 * z_I / (a_Distance + b_Distance)   
k_DC = 4 * z_I / (a_Distance + b_Distance)  

# Calculate distribution factors  
DF_AB = 0.0  # Fixed end  
DF_BA = round(k_BA / (k_BA + k_BC), decimal_point)  
DF_BC = round(k_BC / (k_BA + k_BC), decimal_point)  
DF_CB = round(k_CB / (k_CB + k_CD), decimal_point)  
DF_CD = round(k_CD / (k_CB + k_CD), decimal_point)  
DF_DC = 1.0  

# Moment Distribution Calculation   
M_AB = FEM_AB  
M_BA = FEM_BA  
M_BC = FEM_BC  
M_CB = FEM_CB  
M_CD = FEM_CD  
M_DC = FEM_DC  

# Print Moment Distribution Table Header  
print('')
print("Moment Distribution Table:")  
print("-" * 80) 
print(f"{'Member':<15}{'AB':>10}{'BA':>10}{'BC':>10}{'CB':>10}{'CD':>10}{'DC':>10}")  
print("-" * 80)  
print(f"{'DoF':<15}{round(DF_AB, decimal_point):>10}{round(DF_BA, decimal_point):>10}{round(DF_BC, decimal_point):>10}{round(DF_CB, decimal_point):>10}{round(DF_CD, decimal_point):>10}{round(DF_DC, decimal_point):>10}")   

# Print Initial FEMs  
print(f"{'FEM':<15}{round(M_AB, decimal_point):>10}{round(M_BA, decimal_point):>10}{round(M_BC, decimal_point):>10}{round(M_CB, decimal_point):>10}{round(M_CD, decimal_point):>10}{round(M_DC, decimal_point):>10}")  
print("-" * 80)  

iteration = 0  
while iteration < max_iterations:  
    iteration += 1  

    # Step 1: Balancing moments  
    unbalanced_B = -(M_BA + M_BC)  
    unbalanced_C = -(M_CB + M_CD)  
    unbalanced_D = -(M_DC)  

    bal_BA = unbalanced_B * DF_BA  
    bal_BC = unbalanced_B * DF_BC  
    bal_CB = unbalanced_C * DF_CB  
    bal_CD = unbalanced_C * DF_CD  
    bal_DC = unbalanced_D * DF_DC  
	  
    # Apply balancing moments  
    M_BA += bal_BA  
    M_BC += bal_BC  
    M_CB += bal_CB  
    M_CD += bal_CD  
    M_DC += bal_DC  
	  
    # Print balancing row  
    print(f"{f'Balancing {iteration}':<15}{'-':>10}{round(bal_BA, decimal_point):>10}{round(bal_BC, decimal_point):>10}{round(bal_CB, decimal_point):>10}{round(bal_CD, decimal_point):>10}{round(bal_DC, decimal_point):>10}")  

    # Step 2: Carry-over moments  
    carry_over_AB = bal_BA / 2  
    carry_over_CB = bal_BC / 2  
    carry_over_BC = bal_CB / 2  
    carry_over_CD = bal_DC / 2  
    carry_over_DC = bal_CD / 2  

    # Apply carry-over moments  
    M_AB += carry_over_AB  
    M_CB += carry_over_CB  
    M_BC += carry_over_BC  
    M_CD += carry_over_CD  
    M_DC += carry_over_DC  

    # Calculate the sum of carry-over moments  
    carry_over_sum = abs(carry_over_AB) + abs(carry_over_CB) + abs(carry_over_BC) + abs(carry_over_CD) + abs(carry_over_DC)  

    # Print carry-over row  
    print(f"{f'Carry Over {iteration}':<15}{round(carry_over_AB, decimal_point):>10}{'-':>10}{round(carry_over_BC, decimal_point):>10}{round(carry_over_CB, decimal_point):>10}{round(carry_over_CD, decimal_point):>10}{round(carry_over_DC, decimal_point):>10}")  

    # Check if the sum of carry-over moments is less than carry_over_tolerance  
    if carry_over_sum < carry_over_tolerance:  
        break  

# Print final moments  
print("-" * 80)  
print(f"{'Î£M':<15}{round(M_AB, decimal_point):>10}{round(M_BA, decimal_point):>10}{round(M_BC, decimal_point):>10}{round(M_CB, decimal_point):>10}{round(M_CD, decimal_point):>10}{round(M_DC, decimal_point):>10}")  
print("-" * 80)   

# Calculate reactions at joints  
R_AB = round((a_Load * x_Distance ** 2 / 2 - M_AB - M_BA) / x_Distance, decimal_point)  
R_BA = round(-(a_Load * x_Distance ** 2 / 2 + M_AB + M_BA) / x_Distance, decimal_point)  
if b_Load < c_Load:
    R_BC = round(((b_Load * y_Distance ** 2 / 2) + ((c_Load - b_Load) * y_Distance ** 2 / 2 * 1 / 3) - M_BC - M_CB) / y_Distance, decimal_point)  
    R_CB = round(-((b_Load * y_Distance ** 2 / 2) + ((c_Load - b_Load) * y_Distance ** 2 / 2 * 2 / 3) + M_BC + M_CB) / y_Distance, decimal_point)   
elif b_Load > c_Load:
    R_BC = round(((c_Load * y_Distance ** 2 / 2) + ((b_Load - c_Load) * y_Distance ** 2 / 2 * 2 / 3) - M_BC - M_CB) / y_Distance, decimal_point)  
    R_CB = round(-((c_Load * y_Distance ** 2 / 2) + ((b_Load - c_Load) * y_Distance ** 2 / 2 * 1 / 3) + M_BC + M_CB) / y_Distance, decimal_point)  

R_CD = round((d_Load * b_Distance - M_CD - M_DC)/(b_Distance + a_Distance), decimal_point)  
R_DC = round((d_Load * a_Distance + M_CD + M_DC)/(b_Distance + a_Distance), decimal_point)  

# Print Reactions  
print(f"{'Reaction':<15}{round(R_AB, decimal_point):>10}{round(R_BA, decimal_point):>10}{round(R_BC, decimal_point):>10}{round(R_CB, decimal_point):>10}{round(R_CD, decimal_point):>10}{round(R_DC, decimal_point):>10}")   
print("-" * 80)
```
